## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


## Following are the plots in Seaborn Library:
### Scatter Plot: Show the relationship between two continuous variables using dots. You can create a scatter plot using the scatterplot function.

### Line Plot: Display the trend of a continuous variable over a continuous x-axis. You can create a line plot using the lineplot function.

### Bar Plot: Compare categorical data by displaying bars of different heights. You can create a bar plot using the barplot function.

### Count Plot: Show the count of observations in each category using bars. You can create a count plot using the countplot function.

### Histogram: Display the distribution of a continuous variable using bins. You can create a histogram using the histplot function.

### Box Plot: Visualize the distribution of a continuous variable across different categories using boxes. You can create a box plot using the boxplot function.

### Violin Plot: Similar to a box plot, but also shows the kernel density estimation of the variable. You can create a violin plot using the violinplot function.

### Heatmap: Display the correlation between variables as a colored grid. You can create a heatmap using the heatmap function.

### Pair Plot: Plot pairwise relationships between multiple variables in a dataset. You can create a pair plot using the pairplot function.

### Joint Plot: Visualize the relationship between two variables using a combination of scatter plot and histograms. You can create a joint plot using the jointplot function.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.                                              
## Note: timepoint, signal, event, and region are columns in the fmri dataset.


```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
sns.lineplot(x = fmri.timepoint , y = fmri.signal, hue = fmri.event, style = fmri.region)
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
## Note: pclass, age, and fare are columns in the titanic dataset.


```python
titanic = sns.load_dataset('titanic')
sns.boxplot(x = titanic.pclass ,y = titanic.age)
```

```python
titanic = sns.load_dataset('titanic')
sns.boxplot(x = titanic.pclass ,y = titanic.fare)
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
```

```python
month_to_num = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July':7,
    'August':8,
    'September':9,
    'October':10,
    'November':11,
    'December':12
}
```

```python
flights['month'] = flights['month'].map(month_to_num)
```

```python
sns.heatmap(data = flights)
```
