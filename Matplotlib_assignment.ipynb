## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


### import numpy as np
### np.random.seed(3)
### x = 3 + np.random.normal(0, 2, 50)
### y = 3 + np.random.normal(0, 2, len(x))
### Note: Also add title, xlabel, and ylabel to the plot.

## Ans. Scatter plot: It shows the relationship between two continuous variables by displaying individual data points as dots on a two-dimensional plane.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter PLot')
plt.show()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:


```python
### import numpy as np
### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

## Ans.The subplot function is commonly used in data visualization libraries, such as Matplotlib, to create multiple plots within a single figure. It allows you to divide the figure into a grid of subplots and specify the position of each subplot within that grid.

```python
plt.subplot(2,2,1)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y)
plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)
plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)
plt.subplot(2,2,4)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
## import numpy as np
## company = np.array(["Apple", "Microsoft", "Google", "AMD"])
## profit = np.array([3000, 8000, 1000, 10000])
```

## Ans. A bar plot, also known as a bar chart, is a type of chart that represents categorical data using rectangular bars. It is commonly used to display and compare the values of different categories or groups.

## In a bar plot, the length or height of each bar corresponds to the value or frequency of the category it represents. The bars are typically arranged horizontally or vertically along the axis, depending on the orientation of the chart.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit, color = '#F7D306')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit vs Company chart')
plt.show()
```

```python
plt.barh(company, profit, color = '#F70689')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit vs Company chart')
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
## box1 = np.random.normal(100, 10, 200)
## box2 = np.random.normal(90, 20, 200)
```

## Ans. A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the distribution of a dataset. It displays the range, quartiles, and outliers of the data in a concise and standardized manner.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box 1')
plt.show()
```

```python
plt.boxplot(box2)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box 1')
plt.show()
```
